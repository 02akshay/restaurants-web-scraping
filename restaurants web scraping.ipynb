```python
#import libraryies 
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup #to handel html 
import requests # for scraping
```

```python
columns = ['restaurant', 'type', 'rating', 'delivery_time', 'cost_for_two', 'offer', 'url']
df = pd.DataFrame(columns=columns)
df.head()
```

```python
url = "https://www.XYZ.com/indore?page=1" #you can put i=1 and run complete code in for loop
responce = requests.get(url)
responce
```

##### Response 200 show page successfully scrap
#### You can put URL in for loop so that you can scrape multiplt pages in one command 

```python
data = responce.text #data

soup = BeautifulSoup(data,'html.parser')

soup # To show data in unstructure form 

# soup.prettify # To show data in structure form 
```

```python
main_divs = soup.find_all('div',{'class':'_3XX_A'}) 

for main_div in main_divs:
    hotel_name =  main_div.find(attrs={'class': 'nA6kb'}).text.strip()
    hotel_type =  main_div.find(attrs={'class': '_1gURR'}).text.strip()
     
    details = main_div.find(attrs={'class': '_3Mn31'}).text.split('•')
    rating = details[0]
    delivery_time = details[1]
    apx_cost_for_two = details[2] 
    
    try:
        offer =  main_div.find(attrs={'class': 'sNAfh'}).text
    except:
        offer =  '' #if offers not available

    url =  main_div.find('a').get('href')
    
    # Pass a series in append() to append a row in dataframe  
    series = pd.Series([hotel_name,hotel_type,rating,delivery_time,apx_cost_for_two,offer,url], index=df.columns )
#     print(series)
    df = df.append(series, ignore_index=True)

#     print(main_div.prettify())
    # print('Name ',main_div.get_text(strip=True))
 
```
